```python
import numpy as np
import sys
import math

DNA_SIZE = 3           # DNA length      DNA 長度 維度
DNA_MAX  = 5             # DNA 編碼
POP_SIZE = 10          # population size 人口長度  
CROSS_RATE = 1         # crossover probability 
MUTATION_RATE = 0.2    # mutation probability 
MFE = POP_SIZE  * 1         #計算fintneess 次數
MFE_COUNT = 0
X_BOUND = [-3, DNA_MAX]         # x upper and lower bounds 上下限
pop = np.random.uniform(X_BOUND[0],DNA_MAX , size=(POP_SIZE, DNA_SIZE + 1))   # initialize the pop DNA


def F(x):
    error = 0
    fitness =0     
    y = x[0] * 1
    return y

def getFintness(pop):
    pop = pop[np.lexsort(pop.T)]
    return pop[0,:-1]
      

def select(pop, new):   
    for i,_ in enumerate(new) :
        i_finess = F(_[:-1])
        global MFE_COUNT
        MFE_COUNT = MFE_COUNT + 1
        new[i,-1] = i_finess
    pop = np.row_stack((pop,new))
    pop = pop[np.lexsort(pop.T)]
#     pop_row = pop.shape[0]
#     pop_col = pop.shape[1]
    #pop = pop[pop_row - 1 - POP_SIZE :pop_row - 1, :] # 找最大 排序200 個 由小排到大
    pop = pop[:POP_SIZE,:]
    return pop   
def crossover(pop,copy):
    temp =[0] * (DNA_SIZE + 1)
    for i in range (DNA_SIZE):
        if np.random.rand() < CROSS_RATE:
            r = np.random.rand() 
            i_ = 1 - r 
            x =  np.random.randint(POP_SIZE)
            
            temp[i] = r * pop[i] + i_ * copy[x,i]
    return temp
        
def mutate(child):
    for i in range (DNA_SIZE ):     
        if np.random.rand() < MUTATION_RATE:
            a = np.random.uniform(X_BOUND[0] ,DNA_MAX)
            child[i] = a
    return child

                     
                
def GA():
    temp = np.random.uniform(DNA_MAX, size=(POP_SIZE, DNA_SIZE + 1))
    global pop
    while( MFE_COUNT < MFE):   
        pop = select(pop, temp )
        temp = np.zeros( (1 , DNA_SIZE  + 1 ) )
        new  = pop.copy()   
        for parent in pop:
            child = crossover(parent, new)
            child = mutate(child)
            temp = np.row_stack((temp,child))
        
        temp = np.delete(temp, [0], axis=0)
        print(MFE_COUNT,"最好的基因",getFintness(pop))
        print(MFE_COUNT,"最好的分數",F(getFintness(pop)))
```
